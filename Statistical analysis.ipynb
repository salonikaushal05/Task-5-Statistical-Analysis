# Mean, Median and Mode Analysis

Mean represents the average value of the data.

Median represents the middle value when data is arranged in ascending order.

Mode represents the most frequently occurring value in the dataset.

```python
scores = df["Math_Score"]

print("Mean =", scores.mean())
print("Median =", scores.median())
print("Mode =", scores.mode()[0])
```

# Mean, Median and Mode Analysis

## Results for Math Score

- Mean = 75.20

- Mode = 105.38

## Interpretation

The mode score is 105.38, indicating the most frequently occurring Math Score in the dataset.

## Conclusion

Since the mean and median values are very close, the distribution of Math Scores appears to be fairly balanced. Students achieved an average score of approximately 75 marks in Mathematics.

# T-Test Analysis

A T-Test is performed to determine whether there is a statistically significant difference in Math Scores between students with high attendance and low attendance.

## Hypotheses

### Null Hypothesis (H₀)
There is no significant difference in Math Scores between students with high attendance and low attendance.

### Alternative Hypothesis (H₁)
There is a significant difference in Math Scores between students with high attendance and low attendance.

```python
from scipy.stats import ttest_ind

high_attendance = df[df['Attendance'] >= 85]['Math_Score']
low_attendance = df[df['Attendance'] < 85]['Math_Score']

t_statistic, p_value = ttest_ind(high_attendance, low_attendance)

print("T Statistic =", t_statistic)
print("P Value =", p_value)
```

# T-Test Results

## Results

- T Statistic = -0.4265
- P Value = 0.6698

## Interpretation

The p-value obtained from the T-Test is 0.6698, which is greater than the significance level of 0.05.

Therefore, we fail to reject the Null Hypothesis.

# Conclusion

Based on the statistical analysis:

- Mean, Median, and Mode were calculated for the Math Score.
- A T-Test was performed to compare Math Scores between students with high attendance and low attendance.
- The obtained p-value (0.6698) is greater than 0.05.

Therefore, there is no statistically significant difference in Math Scores between the two attendance groups.
